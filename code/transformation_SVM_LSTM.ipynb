```python
import pandas as pd
from scipy.stats import chi2_contingency
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# Function to calculate Cram√©r's V
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2, _, _, _ = chi2_contingency(confusion_matrix)
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2corr / min((kcorr-1), (rcorr-1)))

```

```python
# Example DataFrame
cat_df = X[categorical_features]

# Calculate correlation matrix using Cram√©r's V
correlation_matrix = cat_df.apply(lambda x: cat_df.apply(lambda y: cramers_v(x, y)))
#print(correlation_matrix)
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='YlGnBu', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix for Categorical Features')
plt.show()
```

# Feature selection: Retain p-significant features

```python
import statsmodels.api as sm

def get_p_significant_features(X,y):
  X = sm.add_constant(X)
  ols_model = sm.OLS(y, X).fit()

  p_features = {}
  for i in X.columns.tolist():
    p_features[f'{i}'] = ols_model.pvalues[i]

  data_p_value= pd.DataFrame(p_features.items(), columns=['Feature_name', 'p-Value']).sort_values(by = 'p-Value').reset_index(drop=True)
  print('-'*50)
  print('p-significant columns/features')
  print('-'*50)
  print(data_p_value[data_p_value['p-Value'] < 0.05])
  significant_features = data_p_value[data_p_value['p-Value']<0.05]['Feature_name'].tolist()

  return significant_features


X_train_significant = get_p_significant_features(X_train_std,Y_train)
print(X_train_significant)
```

#### Evaluation Metrics

```python
from sklearn import metrics
#Evaluation metrics
def metrics_calculation(y_actual, y_predicted):
    
    #Mean average score
    mae =metrics.mean_absolute_error(y_actual, y_predicted)
    
    #Mean Squared Error
    mse=metrics.mean_squared_error(y_actual, y_predicted)
   
    #root mean squared error
    rmse=np.sqrt(mse)
    
    #R2 score
    r2 = metrics.r2_score(y_actual,y_predicted)
    
    results={"MAE":mae,"MSE":mse,"RMSE":rmse,"R2":r2}
    #print(results)
    return results
  
```

#### LSTM Hypertuning

```python
import numpy as np
from sklearn.model_selection import ParameterGrid
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.optimizers import Adam

# Data preparation
X_train_lstm = X_train_h.reshape((X_train_h.shape[0], 1, X_train_h.shape[1]))
X_valid_lstm = X_Valid.reshape((X_Valid.shape[0], 1, X_Valid.shape[1]))

# Hyperparameter grid
param_grid = {
    'units': [50, 100],
    'activation': ['relu', 'tanh'],
    'learning_rate': [0.001, 0.01, 0.1],
    'epochs': [50, 100],
    'batch_size': [32, 64]
}

# Create a function to build the LSTM model
def build_lstm_model(units, activation, learning_rate, epochs, batch_size):
    model = Sequential()
    model.add(LSTM(units=units, activation=activation, input_shape=(X_train_lstm.shape[1], X_train_lstm.shape[2])))
    model.add(Dense(units=1))
    optimizer = Adam(learning_rate=learning_rate)
    model.compile(optimizer=optimizer, loss='mean_squared_error')
    return model

# Create an empty DataFrame to store results
results_df =[]
param=[]
# Grid search
best_loss = np.inf
best_params = None

for params in ParameterGrid(param_grid):
    model = build_lstm_model(**params)
    model.fit(X_train_lstm, Y_train_h, epochs=params['epochs'], batch_size=params['batch_size'], verbose=0)
    loss = model.evaluate(X_valid_lstm, Y_Valid, verbose=0)
    y_pred = model.predict(X_valid_lstm)
    results=metrics_calculation(Y_Valid, y_pred)
    print(f'Test Loss: {loss},params:{params},results:{results}')
    # Store results in DataFrame
    results_df.append(results)
    param.append(params)


    if loss < best_loss:
        best_loss = loss
        best_params = params

print("Best Hyperparameters:", best_params)
print("Best Test Loss:", best_loss)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

X_train_lstm= X_pca_train.reshape((X_pca_train.shape[0], 1, X_pca_train.shape[1]))
X_test_lstm = X_pca_test.reshape((X_pca_test.shape[0], 1, X_pca_test.shape[1]))

# Build the LSTM model
model = Sequential()
model.add(LSTM(units=50, activation='ReLU', input_shape=(X_train_lstm.shape[1], X_train_lstm.shape[2])))
model.add(Dense(units=1))  # Output layer for regression
# Compile the model
model.compile(optimizer=Adam(learning_rate=0.01), loss='mean_squared_error')
# Lists to store metrics for each epoch
mse_list, mae_list, r2_list = [], [], []

# Train the model and record metrics at each epoch
Iterations = 100  # Adjust as needed
for epoch in range(Iterations):
    model.fit(X_train_lstm, Y_train,batch_size=32, verbose=0)  # Train for 1 epoch
    # Evaluate on test data
    y_pred = model.predict(X_test_lstm)
    
    # Calculate metrics
    mse = mean_squared_error(Y_test, y_pred)
    mae = mean_absolute_error(Y_test, y_pred)
    r2 = r2_score(Y_test, y_pred)

    # Append metrics to lists
    mse_list.append(mse)
    mae_list.append(mae)
    r2_list.append(r2)
```

```python
# Plot the metrics vs epoch
epochs_range = range(1, Iterations + 1)

plt.figure(figsize=(12, 4))

# Mean Squared Error plot
plt.subplot(1, 3, 1)
plt.plot(epochs_range, mse_list, label='MSE')
plt.title('Mean Squared Error vs Epoch')
plt.xlabel('Epoch/Iterations')
plt.ylabel('MSE')
plt.legend()

# Mean Absolute Error plot
plt.subplot(1, 3, 2)
plt.plot(epochs_range, mae_list, label='MAE')
plt.title('Mean Absolute Error vs Epoch')
plt.xlabel('Epoch/Iterations')
plt.ylabel('MAE')
plt.legend()

# R-squared plot
plt.subplot(1, 3, 3)
plt.plot(epochs_range, r2_list, label='R2')
plt.title('R-squared vs Epoch')
plt.xlabel('Epoch/Iterations')
plt.ylabel('R2')
plt.legend()

plt.tight_layout()
plt.show()
```
